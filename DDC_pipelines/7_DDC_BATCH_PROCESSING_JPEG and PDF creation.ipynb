```python
#------------------------------------Input data-----------------------------------

# Path address to the folder where RvtExporter.exe|IfcExporter.exe converter is located
path_conv = r'C:\DDC\DDC_Converter\DDC_Converter_Revit_and_IFC_FULL\datadrivenlibs\\'
# Path address of the folder where Revit|IFC files are located
path = r'C:\Revit_Projects\test\\'

#-----------------------------------------------------------------------------------------

import os
import subprocess
import time

def convert_and_wait(path_conv, exporter_name, file_path, extension):
    subprocess.Popen([os.path.join(path_conv, exporter_name), file_path], cwd=path_conv)
    output_file = os.path.join(path, f"{os.path.splitext(file)[0]}_{extension}.xlsx")
    while not os.path.exists(output_file):
        time.sleep(0.5)
    print(f"Conversion Done for Project {os.path.splitext(file)[0]}")  
    
# Ensure output directory exists
os.makedirs(path, exist_ok=True)
# Conversion process from RVT and IFC
for file in os.listdir(path):
    full_path = os.path.join(path, file)
    if file.endswith('.ifc'):
        convert_and_wait(path_conv, 'IfcExporter.exe', full_path, 'ifc')
    elif file.endswith('.rvt'):
        convert_and_wait(path_conv, 'RvtExporter.exe', full_path, 'rvt')
        
        
#-----------------------------------------------------------------------------------------

import os
import pandas as pd
from fpdf import FPDF

import matplotlib.pyplot as plt

def read_excel_file(file_path):
    try:
        wb = openpyxl.load_workbook(file_path, data_only=True)
        sheet = wb.active
        data = [[cell.value for cell in row] for row in sheet]
        wb.close()
        return pd.DataFrame(data)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return pd.DataFrame()  # Return an empty DataFrame in case of error
    
    
def process_xlsx_files(folder_path, s_cat):
    for filename in os.listdir(folder_path):
        if filename.endswith(".xlsx"):
            file_path = os.path.join(folder_path, filename)

            # Load the XLSX file
            df = pd.read_excel(file_path)

            # Handle NA/NaN values in 'Category' column
            df['Category'] = df['Category'].fillna('')  # Replace NA/NaN with empty string

            # Calculate sum of volumes and lengths for the specified category
            dfq = df[df['Category'].str.match(s_cat)]
            dfq = dfq.groupby('Category')['Volume', 'Width'].sum()

            if not dfq.empty:
                cat_len = str(dfq.iloc[0]['Width'])
                cat_vol = str(dfq.iloc[0]['Volume'])
            else:
                cat_len = cat_vol = 'N/A'

            # Inside your loop where you process each file
            plt.figure(figsize=(25, 8))  # Set the desired size here
            pie_chart = df.groupby(['Category']).sum().plot.pie(y='Length', title='Pie Chart: ' + filename)
            plt.savefig(os.path.join(folder_path, filename.replace('.xlsx', '_pie_chart.png')))
            plt.close()  # Close the plot to free up memory

            # Creating a horizontal bar plot
            plt.figure(figsize=(15, 20))  
            barh_plot = df.groupby('Category')['Volume'].sum().plot(kind='barh')
            plt.title('Bar Chart: ' + filename[:-4])
            plt.savefig(os.path.join(folder_path, filename.replace('.xlsx', '_barh_plot.jpeg')))

            # Creating a PDF report
            pdf = FPDF()
            pdf.add_page()
            pdf.set_font('Arial', 'B', 16)
            pdf.cell(190, 8, 'Report for ' + filename[:-4], 2, 1, 'L')
            pdf.set_font('Arial', '', 14)
            pdf.cell(190, 8, 'Sum of volumes for ' + s_cat + ': ' + cat_vol, 2, 1, 'L')
            pdf.cell(190, 8, 'Sum of widths for ' + s_cat + ': ' + cat_len, 2, 1, 'L')

            # Add additional data to the PDF as needed
            pdf.output(os.path.join(folder_path, filename.replace('.xlsx', '_report.pdf')), 'F')

            print(f"Processed {filename}")

# Call the function with the specific category
process_xlsx_files(path, "OST_Walls")

```
